## **outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Tidak ada nilai yang berada di luar batas IQR (Interquartile Range). Dengan kata lain, semua data dalam kolom numerik berada dalam rentang yang dianggap normal menurut metode IQR.

## **Construct Data**

```python
df
```

```python
# Menyimpan daftar kolom awal sebelum modifikasi
original_columns = df.columns.tolist()

# Fungsi untuk menentukan kategori pengguna berdasarkan Daily_Screen_Time_Hours
def categorize_user(screen_time_hours):
    if screen_time_hours < 4:
        return "Light User"
    elif 4 <= screen_time_hours <= 8:
        return "Moderate User"
    else:
        return "Heavy User"

# Menambahkan kolom 'User_Category' tanpa mengubah struktur dataset lainnya
df["User_Category"] = df["Daily_Screen_Time_Hours"].apply(categorize_user)

df = df[original_columns + ["User_Category"]]  # menambahn 'User_Category'

# Mengecek bentuk akhir dataset
df.shape
```

User_Category berhasil di tambahkan berdasarkan Daily_Screen_Time_Hours

### 1. Heatmap Korelasi Antar Variabel Numerik
Insight:

- Tidak ada korelasi yang kuat antara variabel-variabel yang dianalisis.
- Waktu penggunaan aplikasi sosial media, gaming, dan produktivitas memiliki korelasi yang sangat kecil terhadap jumlah total aplikasi yang digunakan dan waktu layar harian.

### 2. Jumlah Pengguna per Lokasi
Insight:

- New York memiliki jumlah pengguna terbanyak dibandingkan kota lainnya.
- Perbedaan jumlah pengguna antar kota tidak terlalu signifikan.

```python
# Visualisasi jumlah pengguna per lokasi (menggunakan kolom 'Location')
plt.figure(figsize=(8, 5))
plt.bar(df['Location'].value_counts().index, df['Location'].value_counts().values, color='pink')
plt.title('Jumlah Pengguna per Lokasi')
plt.xlabel('Lokasi')
plt.ylabel('Jumlah Pengguna')
plt.xticks(rotation=45)
plt.show()
```

### 3. Distribusi Waktu Penggunaan Aplikasi
Insight:

- Penggunaan aplikasi sosial media, gaming, dan produktivitas memiliki proporsi yang hampir sama.
- Tidak ada dominasi yang terlalu besar dari satu kategori aplikasi tertentu.

```python
# Pie Chart (Matplotlib): Distribusi Waktu Penggunaan Aplikasi
plt.figure(figsize=(6, 6))
plt.pie(df[['Social_Media_Usage_Hours', 'Productivity_App_Usage_Hours', 'Gaming_App_Usage_Hours']].sum(),
        labels=['Media Sosial', 'Produktivitas', 'Gaming'],
        autopct='%1.1f%%',
        colors=['lightcoral', 'lightblue', 'lightgreen'])

plt.title('Distribusi Waktu Penggunaan Aplikasi')
plt.show()
```

### 4. Rata-rata Jumlah Aplikasi yang Digunakan Berdasarkan Usia
Insight:

- Penggunaan aplikasi bervariasi berdasarkan usia, tetapi tidak menunjukkan pola yang sangat jelas.
- Kelompok usia tertentu memiliki jumlah aplikasi yang digunakan lebih tinggi daripada yang lain.

```python
# Mengelompokkan jumlah aplikasi yang digunakan berdasarkan usia (hanya usia yang ada di dataset)
df_age_group = df.groupby('Age')['Total_App_Usage_Hours'].mean()

# Visualisasi jumlah aplikasi yang digunakan berdasarkan usia
plt.figure(figsize=(10, 5))
plt.plot(df_age_group.index, df_age_group.values, marker='o', linestyle='-', color='red')
plt.title('Rata-rata Jumlah Aplikasi yang Digunakan Berdasarkan Usia')
plt.xlabel('Usia')
plt.ylabel('Rata-rata Jumlah Aplikasi yang Digunakan')
plt.grid()
plt.show()
```
